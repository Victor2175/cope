# Load the observational data with 5 degrees granularity

```python
def load_obs():
    """
    Load centered observational data.
    """
    # define the file
    file = '/home/simondi/simondi/cope-analysis/data/erss/sst_annual_g050_mean_19812014_centered.nc'

    # read the dataset
    file2read = netcdf.Dataset(file,'r')
    file_split = file.split("_")

    # load longitude, latitude and sst monthly means
    lon = np.array(file2read.variables['lon'][:])
    lat = np.array(file2read.variables['lat'][:])
    sst = np.array(file2read.variables['sst'])
    
    
    return lon, lat, sst
```

# Define test set

```python
test_set = 0
test_model = 'ACCESS-ESM1-5'
idx = 0
for idx_run, (run, run_arr) in enumerate(dic_model_forcing[test_model]['ssp585'].items()):
    print(idx_run)
    if idx_run ==0:
        test_set = dic_model_forcing[test_model]['ssp585'][run]/len(dic_model_forcing[test_model]['ssp585'])
    else:
        test_set = np.add(test_set, dic_model_forcing[test_model]['ssp585'][run]/len(dic_model_forcing[test_model]['ssp585']))
```

```python
import sklearn
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# reshape the arrays 
train_set_tmp = scaler.fit_transform(np.reshape(train_set,(train_set.shape[0], train_set.shape[1]*train_set.shape[2])))
test_set_tmp = scaler.transform(np.reshape(test_set,(test_set.shape[0], test_set.shape[1]*test_set.shape[2])))

# define ridge regression problem
model  = sklearn.linear_model.Ridge(alpha=1.0)

Y = np.nan_to_num(np.nanmean(train_set_tmp,axis=1))
X = np.nan_to_num(train_set_tmp)
# test_piControl

model.fit(X, Y)
```
